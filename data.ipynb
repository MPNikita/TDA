```python
import pandas as pd
import pickle
import os

# Путь к папке с данными
data_folder = 'data'

# Названия файлов
file_names = [
    'adjectives100_ruscorp_ppmi_300.pkl',
    'adjectives100_ruscorp_add_ppmi_300.pkl'
]

# Читаем каждый файл и выводим его содержимое
for file_name in file_names:
    file_path = os.path.join(data_folder, file_name)
    
    # Открываем .pkl файл с указанием кодировки
    try:
        with open(file_path, 'rb') as f:
            data = pickle.load(f, encoding='latin1')  # Попробуйте 'bytes', если 'latin1' не работает
    except Exception as e:
        print(f"Ошибка при загрузке файла {file_name}: {e}")
        continue
    
    # Преобразуем данные в DataFrame (если это возможно)
    if isinstance(data, (dict, list)):
        df = pd.DataFrame(data)
    else:
        df = pd.Series(data).to_frame()  # Если данные не являются словарем или списком
    
    # Выводим название файла и данные
    print(f"Файл: {file_name}")
    display(df)  # Используем display для красивого вывода в Jupyter Notebook
    print("\n" + "="*50 + "\n")  # Разделитель между файлами
```

```python
import pandas as pd
import pickle
import os
import warnings

# Отключаем предупреждения для чистоты вывода
warnings.filterwarnings('ignore')

# Проверяем существование папки data
if not os.path.exists('data'):
    raise FileNotFoundError("Папка 'data' не найдена в текущей директории!")

# Список целевых файлов
files = [
    'adjectives100_ruscorp_ppmi_300.pkl',
    'adjectives100_ruscorp_add_ppmi_300.pkl'
]

def load_pkl_safe(file_path):
    """Безопасная загрузка pickle-файла с обработкой разных кодировок"""
    encodings = ['latin1', 'bytes', 'utf8']
    for encoding in encodings:
        try:
            with open(file_path, 'rb') as f:
                return pickle.load(f, encoding=encoding)
        except Exception as e:
            last_error = e
    raise last_error

# Создаем HTML-таблицу для отчета
html_output = ['<h2>Анализ pickle-файлов</h2>']

for file_name in files:
    file_path = os.path.join('data', file_name)
    
    # Проверка существования файла
    if not os.path.exists(file_path):
        html_output.append(f'<div style="color: red;">Файл {file_name} не найден!</div>')
        continue
    
    # Пытаемся загрузить данные
    try:
        data = load_pkl_safe(file_path)
        html_output.append(f'<h3>Файл: {file_name}</h3>')
        
        # Анализ структуры данных
        if isinstance(data, dict):
            df = pd.DataFrame.from_dict(data, orient='index').T
        elif isinstance(data, (list, tuple)):
            df = pd.DataFrame(data)
        else:
            df = pd.Series(data).to_frame()
            
        # Генерация HTML-представления
        html_output.append(df.style.set_table_attributes('class="dataframe"').render())
        
    except ModuleNotFoundError as e:
        html_output.append(f'''
            <div style="color: red; border: 1px solid #ff0000; padding: 10px;">
                <b>Критическая ошибка:</b> {e}<br>
                Попробуйте установить недостающий модуль:<br>
                <code>!pip install {e.name}</code>
            </div>
        ''')
    except Exception as e:
        html_output.append(f'<div style="color: red;">Ошибка: {str(e)}</div>')

    html_output.append('<hr>')

# Выводим итоговый отчет
from IPython.display import HTML, display
display(HTML('\n'.join(html_output)))

# Дополнительный анализ через pickletools (по требованию)
if 'show_pickle_structure' in locals():
    import pickletools
    print('\n\nСтруктура первого файла:')
    with open(os.path.join('data', files[0]), 'rb') as f:
        pickletools.dis(f)
```

```python
import pickle
import os

# Пути к файлам
file_paths = [
    os.path.join('data', 'adjectives100_ruscorp_ppmi_300.pkl'),
    os.path.join('data', 'adjectives100_ruscorp_add_ppmi_300.pkl')
]

# Словарь для хранения данных
loaded_data = {}

for path in file_paths:
    try:
        with open(path, 'rb') as f:
            loaded_data[path] = pickle.load(f)
        print(f"✓ Файл {os.path.basename(path)} успешно загружен")
        
    except Exception as e:
        print(f"✕ Ошибка при загрузке {os.path.basename(path)}: {str(e)}")
        loaded_data[path] = None

# Анализ загруженных данных
print("\nАнализ данных:")
for path, data in loaded_data.items():
    if data is not None:
        print(f"\nФайл: {os.path.basename(path)}")
        print(f"Тип данных: {type(data)}")
        
        if isinstance(data, dict):
            print(f"Количество ключей: {len(data)}")
            print("Пример ключей:", list(data.keys())[:3])
            
        elif isinstance(data, (list, tuple)):
            print(f"Количество элементов: {len(data)}")
            print("Первый элемент:", data[0] if len(data) > 0 else "Нет данных")
            
        elif hasattr(data, 'shape'):
            print(f"Размерность данных: {data.shape}")
            
        else:
            print("Неизвестный формат данных")
```
